```python
import pandas as pd
import numpy as np
```

```python
bill = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ANN/Real_Combine.csv')
bill.head()
```

```python
bill = bill.dropna()
```

```python
bill.describe()
```

```python
x = bill.iloc[:,:8]
y = bill.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 100)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, ReLU, ELU, Dropout
```

```python
regressor = Sequential()
```

```python
regressor.add(Dense(units=5, kernel_initializer='he_uniform', activation='relu', input_dim = 8))
```

```python
regressor.add(Dense(units=3, kernel_initializer='he_uniform', activation='relu'))
```

```python
regressor.add(Dense(units=1, activation='linear'))
```

```python
regressor.summary()
```

```python
regressor.compile(optimizer='adam', loss='mae', metrics=['mae'])
```

```python
model_train = regressor.fit(x_train, y_train, batch_size=10, epochs=100)
```

```python
predict = regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
print(mean_absolute_error(y_test, predict))
print(r2_score(y_test, predict))
```
